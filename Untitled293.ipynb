```python
df = df.drop('Unnamed: 0',1)
df = df.drop('emp_id',1)
df = df.drop('recruitment_type',1)
df = df.drop('Dept',1)
df = df.drop('location',1)
```

```python
 def education_to_numeric(x):
        if x=='PG': return 1
        if x=='UG':   return 2

df['education'] = df['education'].apply(education_to_numeric)
print(df)
```
